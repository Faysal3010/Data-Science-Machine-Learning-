```python
import matplotlib.pyplot as plt
```

```python
import numpy as np # Import numpy here as well to ensure it's available

x=np.array([1,2,2.5,6,3,6,8])
y=np.array([20,50,55,90,30,90,120])
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0] # Number of training examples
    cost = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i])**2

    total_cost = cost / (2 * m) # Average cost (MSE / 2)

    return total_cost
```

```python
w,b=float(input("Enter w: ")),float(input("Enter b: "))
plt.scatter(x,y,c='r',marker='x', label='Original Data')
plt.plot(x, [w*xi + b for xi in x], label=f'y = {w}x + {b}', color='blue') # Add the line

# Plot vertical lines from each scatter point to the fitted line
for i in range(len(x)):
    y_hat_i = w * x[i] + b
    plt.vlines(x[i], min(y[i], y_hat_i),max(y[i], y_hat_i), color='purple', linestyle='dotted')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Original Data with Linear Fit and Residuals')
plt.legend()
plt.grid(True)
plt.show()
print(compute_cost(x, y, w, b))
```
